```python
import torch
```

```python
import nltk
```

```python
f = [i.strip() for i in f if i.strip() != ''] #32 is emprical, removing the table of contents because it sucked.
f = [i.lower() for i in f]
```

```python
def clean(s : str):
    return ''.join([i for i in list(s) if i.isalpha() or i == ' '])
```

```python
f = [clean(i) for i in f]
```

```python
f.pop(0)
```

```python
words = [[i for i in j.split(' ')] for j in f]
```

```python
len(words)
```

```python
for i in range(32000):
  words[i]=['.'] + words[i] + ['.']
```

```python
b={}
for i in range(32000):
  sentence=words[i]
  for w1,w2 in zip(sentence,sentence[1:]):
    bigram=(w1,w2)
    b[bigram]=b.get(bigram,0)+1
```

```python
words2 = [set([i for i in j.split(' ')]) for j in f]
```

```python
vocab = []
for i in words2:
    vocab += list(i)
vocab = list(set(vocab))
```

```python
len(vocab)
```

```python
N = torch.zeros((24000, 24000), dtype=torch.int32)
```

```python
stoi = {s : i+1 for i,s in enumerate(vocab)}
stoi['.'] = 0 #special character
itos = {i:s for s,i in stoi.items()}
```

```python
for i in range(32000):
  sentence=words[i]
  for w1,w2 in zip(sentence,sentence[1:]):
    ix1 = stoi[w1]
    ix2 = stoi[w2]
    N[ix1, ix2] += 1
```

```python
P = (N+1).float()
P /= P.sum(1, keepdims=True)
```

```python
g = torch.Generator().manual_seed(2147483649)

for i in range(5):

  out = []
  ix = 0
  while True:
    p = P[ix]
    ix = torch.multinomial(p, num_samples=1, replacement=True, generator=g).item()
    out.append(itos[ix])
    if ix == 0:
      break
  print(out)
```
